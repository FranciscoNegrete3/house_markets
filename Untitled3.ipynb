```python
import pandas as pd
import time
import warnings
```

```python
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
url = "https://redfin-public-data.s3.us-west-2.amazonaws.com/redfin_market_tracker/city_market_tracker.tsv000.gz"
```

```python
# start time to read file
startTime = time.time()
df = pd.read_csv(url, compression='gzip', sep='\t') #, on_bad_lines='skip')
# end time 
executionTime = (time.time() - startTime)
print('Execution time in minutes: ' + str(round(executionTime / 60, 2)))
print('Num of rows:', len(df))
print('Num of cols:', len(df.columns))

df.tail()
```
